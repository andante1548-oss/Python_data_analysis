```python
class Calculator:
  def setData(self, first, second):
    self.first = first #가독성 확보를 위해 이름을 동일하게 유지함
    self.second = second
  def add(self): #메서드
    self.result = self.first + self.second
  def div(self):
    self.result = self.first / self.second

a = Calculator()
a.setData(10,20)
a.add()
print(a.first)
print(a.second)

b = Calculator()
b.setData(30,40)
b.add()
print(b.first)
print(b.second)
print(a.first)
```

```python
class Calculator:
  def __init__(self, first, second): ##클래스변수 초기화 : 생성자 = constructor
    self.first = first
    self.second = second
  def add(self): #메서드
    self.result = self.first + self.second
  def div(self):
    self.result = self.first / self.second

a = Calculator(10, 20) #생성자에 값을 주기 위하여
print(a.first)
print(a.second)

b = Calculator(30, 40)
print(b.first)
print(b.second)
```

```python
class FourCal(Calculator):
  def mul(self):
    self.result = self.first * self.second
  def sub(self):
    self.result = self.first - self.second
  #상속
calc = FourCal(10, 20)
calc.add()
print(calc.result)
```

```python
a = FourCal(0,15)
a.div()
print(a.result)

#상속해준 클래스 - 부모클래스 / 상속 받은 클래스 - 자식 클래스
class FourCal(Calculator):
  def mul(self):
    self.result = self.first * self.second
  def sub(self):
    self.result = self.first - self.second
  def div(self) : #override  == 재정의
    if self.second == 0:
      self.result =0
    else :
      self.result = self.first / self.second

a = FourCal(0,15)
a.add()
```
