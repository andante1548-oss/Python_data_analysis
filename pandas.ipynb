```python
obj3 = pd.Series(["blue", "purple", "yellow"], index=[0, 2, 4])
print(obj3)
reindexed = obj3.reindex(np.arange(6), method="ffill") #앞의값을 뒷값으로 채움
print(reindexed)
reindexed = obj3.reindex(np.arange(6), method="bfill") #뒷값을 앞값으로 채움
print(reindexed)
```

```python
#Dataframe
data = pd.DataFrame(np.arange(16).reshape((4, 4)),
                    index=["Ohio", "Colorado", "Utah", "New York"],
                    columns=["one", "two", "three", "four"])
print(data)
newdata = data.drop(index=["Colorado", "Ohio"])
print(newdata)
newdata = data.drop(columns=["two"])
print(newdata)
newdata = data.drop("three", axis=1) #axis=1 : 열 (행기준, y축 기준)
print(newdata)
newdata = data.drop("four", axis="columns")
print(newdata)
newdata = data.drop("Ohio", axis=0) #axis=0 : 행 (열기준, x축 기준)
print(newdata)
newdata = data.drop("Colorado", axis="index") #axis=0 : 행 (열기준, x축 기준)
print(newdata)
```

```python
#Indexing, Selection, Filtering
obj = pd.Series(np.arange(4.), index=["a", "b", "c", "d"])
print(obj)
print(obj["b"])
print(obj[1])
print(obj[2:4])
print(obj[["b", "a", "d"]])
print(obj[[1, 3]])
print(obj[obj < 2])
print(obj.loc[["b", "a", "d"]])
print(obj.iloc[[1,0,2]])
print(obj.loc["b": "c"])
print(obj.iloc[1:3]) #암시적 index와 명시적 index의 범위가 다름
obj.loc["b": "c"] = 5
print(obj)
```

```python
data = pd.DataFrame(np.arange(16).reshape((4, 4)),
                    index=["Ohio", "Colorado", "Utah", "New York"],
                    columns=["one", "two", "three", "four"])
print(data)
print(data["two"])
print(data[["three", "one"]])
print(data[:2]) ; print(data.iloc[:2])  #암시적 index를 사용하여 행을 가져오는 것이 가능
print(data[data["three"] > 5]) #bool index
data[data<5] = 0
print(data)
print(data.loc["Colorado"])
print(data.loc[["Colorado", "New York"]])
print(data.loc["Colorado"],["two", "three"])

print(data.iloc[2]) # 암시적 사용
print(data.iloc[[2, 1]])
print(data.iloc[2, [3, 0, 1]])
print(data.iloc[[1, 2], [3, 0, 1]])
print(data.iloc[:, :3][data.three > 5])
data.loc[data.three >= 2]
```
