```python
name = "ㅇ"
age = 30
height = 162.2
print("{0}의 나이는 {1}이고 키는 {2}입니다".format(name, age, height))
# format 함수를 사용하는 경우 입력 순서와 인자가 일치하지 않아도 출력 가능
print("{0}의 키는 {2}이고 나이는 {1}입니다".format(name, age, height))
print("{0}의 나이는 {1}이고 키는 {2}입니다".format(name, age, height)) #앞번호에 인자의 이름을 부여하는 경우 뒷 번호는 자동으로 인자 사용이 불가함 >> 인자의 이름은 뒤에서부터 순서대로 부여할 것
print("{0:10}의 나이는 {1}이고 키는 {2}입니다".format(name, age, height))
print("{0:>10}의 나이는 {1:>10}이고 키는 {2:>10}입니다".format(name, age, height))
print("{0:<10}의 나이는 {1:<10}이고 키는 {2:<10}입니다".format(name, age, height))
print("{0:^10}의 나이는 {1:^10}이고 키는 {2:^10}입니다".format(name, age, height))
print("{0:-^10}의 나이는 {1:*^10}이고 키는 {2:@^10}입니다".format(name, age, height))
```
