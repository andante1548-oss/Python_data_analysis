```python
import os
import re
import socket
import urllib.request
from contextlib import closing
from time import sleep

import pandas as pd
from bs4 import BeautifulSoup

# ===== 사용자 설정 =====

# 기본 URL (페이지 번호 p만 바꿔서 순회)
# ✅ p=1로 고정하지 말고, {page} 자리만 바꿔 쓰도록 수정
BASE_URL = "https://steamcommunity.com/app/2246340/reviews/?p={page}&browsefilter=mostrecent"

# 저장 경로 (원하시는 경로)
OUTPUT_DIR = r"C:\Users\ekgus\OneDrive\documents\MHW_project"
OUTPUT_CSV = os.path.join(OUTPUT_DIR, "monster_hunter_wilds_reviews.csv")

# HTTP 관련 설정
TIMEOUT = 30          # 요청 타임아웃 (초)
MAX_RETRIES = 3       # 최대 재시도 횟수
PAUSE = 1.0           # 요청 간 대기 시간 (초)

# 크롤링할 최대 페이지 수 (예: 5페이지까지; 필요에 따라 조정)
MAX_PAGE = 1

# ===== 공통 함수들 =====

def download_page(url, maxretries=MAX_RETRIES, timeout=TIMEOUT, pause=PAUSE):
    """
    지정한 URL에서 HTML 페이지를 가져오는 함수.
    실패 시 maxretries 만큼 재시도.
    """
    tries = 0
    htmlpage = None
    while tries < maxretries and htmlpage is None:
        try:
            with closing(urllib.request.urlopen(url, timeout=timeout)) as f:
                htmlpage = f.read()
                sleep(pause)
        except (urllib.error.URLError, socket.timeout, socket.error) as e:
            print(f"다운로드 에러 ({e}), 재시도 중... ({tries+1}/{maxretries})")
            tries += 1
    return htmlpage


def parse_reviews_from_html(html, start_serial=1):
    """
    하나의 리뷰 페이지 HTML에서
    [일련번호, 추천/비추천, 플레이 시간, 게시 일시, 리뷰 본문]을 파싱.
    start_serial: 전체 일련번호 시작값 (이전 페이지까지의 개수 + 1)
    """
    soup = BeautifulSoup(html, "html.parser")

    results = []
    serial = start_serial

    # 1) 리뷰 블록 선택
    # 질문에서 주신 HTML 구조에 맞게 셀렉터 사용
    # <div class="apphub_UserReviewCardContent"> ... </div>
    review_blocks = soup.select("div.apphub_UserReviewCardContent")

    for block in review_blocks:
        # 2) 추천/비추천 (예: "추천", "비추천")
        rec_tag = block.select_one("div.title")
        rec_text = rec_tag.get_text(strip=True) if rec_tag else ""

        # 3) 플레이 시간 (예: "기록상 305.1시간")
        playtime_tag = block.select_one("div.hours")
        playtime_text = playtime_tag.get_text(strip=True) if playtime_tag else ""

        # 4) 게시 일시 (예: "게시 일시: 2026년 2월 4일")
        date_tag = block.select_one("div.date_posted")
        date_text = date_tag.get_text(strip=True) if date_tag else ""

        # 5) 리뷰 본문
        # <div class="apphub_CardTextContent"> 안의 텍스트 전체 사용
        content_container = block.select_one("div.apphub_CardTextContent")
        if content_container:
            # date_posted 내부 텍스트는 이미 위에서 따로 뽑았으므로,
            # 필요하다면 제거하고 순수 본문만 남길 수도 있음
            # 여기서는 전체 텍스트에서 날짜 부분만 제거하는 방식 사용
            raw_text = content_container.get_text(separator=" ", strip=True)
            raw_text = re.sub(r"\s+", " ", raw_text)

            if date_text:
                content_text = raw_text.replace(date_text, "").strip()
            else:
                content_text = raw_text
        else:
            content_text = ""

        results.append({
            "일련번호": serial,
            "추천": rec_text,
            "플레이시간": playtime_text,
            "게시일시": date_text,
            "리뷰본문": content_text
        })
        serial += 1

    return results, serial


def crawl_monster_hunter_wilds(max_page=MAX_PAGE):
    """
    Monster Hunter Wilds(2246340)의 리뷰를
    1페이지부터 max_page까지 순회하며 수집.
    """
    all_reviews = []
    serial = 1

    for page in range(1, max_page + 1):
        url = BASE_URL.format(page=page)
        print(f"페이지 {page} 수집 중: {url}")

        html_bytes = download_page(url)
        if html_bytes is None:
            print(f"페이지 {page} 다운로드 실패, 계속 진행합니다.")
            continue

        html = html_bytes.decode("utf-8", errors="ignore")

        page_reviews, serial = parse_reviews_from_html(html, start_serial=serial)
        print(f"페이지 {page}에서 {len(page_reviews)}개 리뷰 수집")

        if not page_reviews:
            # 더 이상 리뷰가 없으면 중단
            print("더 이상 리뷰가 없습니다. 수집을 종료합니다.")
            break

        all_reviews.extend(page_reviews)

    return all_reviews


def main():
    print(f"현재 작업 디렉토리: {os.getcwd()}")
    print(f"출력 디렉토리: {OUTPUT_DIR}")
    print(f"CSV 저장 경로: {OUTPUT_CSV}")

    # 저장 폴더가 없으면 생성
    if not os.path.exists(OUTPUT_DIR):
        os.makedirs(OUTPUT_DIR, exist_ok=True)
        print(f"출력 디렉토리를 생성했습니다: {OUTPUT_DIR}")

    reviews = crawl_monster_hunter_wilds(max_page=MAX_PAGE)

    if not reviews:
        print("수집된 리뷰가 없습니다. CSV 파일을 생성하지 않습니다.")
        return

    df = pd.DataFrame(reviews, columns=["일련번호", "추천", "플레이시간", "게시일시", "리뷰본문"])

    # 1) 파이썬 내에서 DataFrame 내용 확인
    print("\n=== 수집된 리뷰 DataFrame (상위 5개) ===")
    print(df.head())
    print("\n총 행 수:", len(df))

    # 2) CSV 파일로 저장
    try:
        df.to_csv(OUTPUT_CSV, index=False, encoding="utf-8-sig")
        print(f"\nCSV 저장 시도 완료: {OUTPUT_CSV}")
    except Exception as e:
        print(f"\nCSV 저장 중 에러 발생: {e}")
        return

    # 3) 실제로 파일이 존재하는지 확인
    if os.path.exists(OUTPUT_CSV):
        print(f"✅ CSV 파일이 정상적으로 저장되었습니다: {OUTPUT_CSV}")
    else:
        print(f"⚠ CSV 파일이 경로에 보이지 않습니다: {OUTPUT_CSV}")
        print("다음 사항을 확인해 보세요:")
        print("- 경로 문자열에 오타가 없는지")
        print("- OneDrive 동기화/권한 문제는 없는지")
        print("- 관리자 권한이 필요한 폴더는 아닌지")


if __name__ == "__main__":
    main()
```
